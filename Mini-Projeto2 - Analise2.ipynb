```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
list(df)
```

## Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a ca
plt.figure(figsize=(16,9))
plot = sns.countplot(data=df,y='brand',palette='PuOr')
plt.xlabel('Quantidade')
plt.ylabel('Marca')
plt.title('Quantidade de veículos por marca')
plt.show()
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio
plt.figure(figsize=(16,9))
sns.barplot(data=df,x='vehicleType',y='price',hue='gearbox',palette='cool')
plt.xlabel('Tipo de veículo', fontsize=15)
plt.ylabel('Preço',fontsize=15)
plt.title('Preço médio pelo tipo de véiculo e transmissão')
```
