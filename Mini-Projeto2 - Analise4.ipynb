```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
list(df)
```

```python
# Calcule a média de preço por marca e por tipo de veículo
df_copy = df.copy()
```

```python
avg = df_copy.groupby(by=['brand','vehicleType'])['price'].mean()
```

```python
frame = avg.to_frame()
```

```python
framedf = frame.reset_index()
```

```python
list(framedf)
```

```python
framedf.round({'price':2})
frame.head(10)
```

## Preço médio de um veículo por marca, bem como tipo de veículo

```python
# Crie um Heatmap com Preço médio de um veículo por marca, bem como tipo de veículo
pivot = framedf.pivot('brand','vehicleType','price')
plt.figure(figsize=(16,20))
sns.heatmap(data=pivot,annot=True,fmt='.2f',cmap='PuBu')
plt.show()
```
