## Análise 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Distribuição de Veículos com base no Ano de Registro

```python
list(df)
```

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
fig = plt.figure(figsize=(16,9))
sns.histplot(df['yearOfRegistration'],kde=True)
plt.xlabel('Ano de registro', fontsize = 15)
plt.ylabel('Quantidade & KDE', fontsize = 15)
plt.title('Distribuição de Veículos com base no Ano de Registro', fontsize=15)
plt.show()


```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers
fig = plt.figure(figsize=(16,9))
sns.boxplot(data=df,x='vehicleType',y='price')
plt.xlabel('Tipo de veículo', fontsize=15)
plt.ylabel('Preço', fontsize=15)
plt.title('Faixa de preço por tipo de veículo', fontsize=15)

plt.show()
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria 
fig = plt.figure(figsize=(16,9))
sns.countplot(data=df,x='vehicleType')
plt.xlabel('Tipo de veículo', fontsize=15)
plt.ylabel('Quantidade', fontsize=15)
plt.title('Quantidade por tipo de veículo', fontsize=15)

plt.show()
```
