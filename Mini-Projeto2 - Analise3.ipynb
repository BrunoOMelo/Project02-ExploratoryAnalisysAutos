```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
list(df)
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
plt.figure(figsize=(16,9))
sns.barplot(data=df,x='fuelType',y='price',hue='gearbox')
plt.xlabel('Tipo de Combustível', fontsize=15)
plt.ylabel('Preço', fontsize=15)
plt.title('Preço médios dos veículos pelo combustível e transmissão',fontsize=15)
plt.show()
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
plt.figure(figsize=(16,9))
sns.barplot(data=df,x='vehicleType',y='powerPS',hue='gearbox')
plt.xlabel('Tipo de veículo', fontsize=15)
plt.ylabel('Preço', fontsize=15)
plt.title('Preço do tipo de veículo por tipo de trasnmissão',fontsize=15)
plt.show()
```
